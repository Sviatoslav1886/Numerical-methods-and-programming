# Using Python, build a set of solutions (polygon) of the system of linear constraints-inequalities on the plane and geometrically find the largest and smallest values of the linear function z in this polygon

![alt text](photo_2024-07-21_13-16-49.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt

 
def inequality1(x):
    return 4 - x

def inequality2(x):
    return (8 + x) / 2

def inequality3(x):
    return 2 * x - 4

# Equations of the plane
def plane_eq(x):
    return ((-3 * x ) / 2) + 5  
 
x_values = np.linspace(-6, 6, 10)

plt.plot(x_values, inequality1(x_values), label=r'$x_{1}+x_{2}\geq4$')
plt.plot(x_values, inequality2(x_values), label=r'$-x_{1}+2x_{2}\leq8$')
plt.plot(x_values, inequality3(x_values), label=r'$2x_{1}-x_{2}\leq4$')
plt.plot(x_values, plane_eq(x_values),color="black")

# Normal vector
normal_vector = np.array([3, 2])
plt.xlim(-6, 6)
plt.ylim(-6, 10)
 
origin = [0, 0]
plt.quiver(*origin, *normal_vector, color='r', scale=30)   
 
plt.xlabel('$x_{1}$')
plt.ylabel('$x_{2}$')
plt.title(' ')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def inequality1(x):
    return (2 * x + 5) / 3

def inequality2(x):
    return (x - 2) / 2

def inequality3(x):
    return 6 - x

x_values = np.linspace(-50, 50, 100)

# Construction of the first inequality
plt.figure(figsize=(8, 6))
plt.plot(x_values, inequality1(x_values), label=r'$x_{1}+x_{2}\geq4$')
plt.xlabel('$x_{1}$')
plt.ylabel('$x_{2}$')
plt.title('Inequality 1')
plt.legend()
plt.grid(True)
plt.show()

# Construction of the second inequality
plt.figure(figsize=(8, 6))
plt.plot(x_values, inequality2(x_values), label=r'$-x_{1}+2x_{2}\leq8$')
plt.xlabel('$x_{1}$')
plt.ylabel('$x_{2}$')
plt.title('Inequality 2')
plt.legend()
plt.grid(True)
plt.show()

# Construction of the third inequality
plt.figure(figsize=(8, 6))
plt.plot(x_values, inequality3(x_values), label=r'$2x_{1}-x_{2}\leq4$')
plt.xlabel('$x_{1}$')
plt.ylabel('$x_{2}$')
plt.title('Inequality 3')
plt.legend()
plt.grid(True)
plt.show()
```
