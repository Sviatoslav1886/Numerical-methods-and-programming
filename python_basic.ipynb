## <b><span style='color:#F1C40F'>|</span> Task 2</b>

```python
def square_non_negative_numbers(a, b, c):
    if a >= 0:
        a_squared = a ** 2
        print("Квадрат числа a: ", a_squared)
    if b >= 0:
        b_squared = b ** 2
        print("Квадрат числа b: ", b_squared)
    if c >= 0:
        c_squared = c ** 2
        print("Квадрат числа c: ", c_squared)
```

```python
a = 2.0
b = -4.5
c = 3.0

print(a, b, c, '\n')
square_non_negative_numbers(a, b, c)
```

### <b><span style='color:#F1C40F'>|</span> Task 3.1</b>

```python
def task_3_14(x, a, n):
    result = (x + a)**2
    i = n-1

    while i!=0:
        result = (result + a)**2 
        i -= 1
    result += a
    return result
```

```python
x = 1.0
a = 2.0
n = 1

print('x: ', x)
print('a: ', a)
print('n: ', n)

result = task_3_14(x, a, n)
print("\n", "Результат обчислення виразу:", result)
```

### <b><span style='color:#F1C40F'>|</span> Task 3.2</b>

```python
def task_3_24(x, n):
    divisible = 1            # Ділене
    divisor = 1              # Дільник

    # Обчислення чисельника
    for i in range(n*2):
        divisible *= x

    # Обчислення знаменика
    for i in range(n):
        divisor *= 2

    # Виконання ділення
    result = divisible / divisor

    return result
```

```python
# n = int(input("Введіть натуральне число n: "))
# x = float(input("Введіть дійсне число x: "))
n = 2
x = 3

print("n:", n)
print("x:", x)
task_3_24(n, x)
```

### <b><span style='color:#F1C40F'>|</span> Task 4.1</b>

```python
def func_6_d(s):
    for i in range(len(s) - 1):
        if (s[i:i+2] == 'di' or s[i:i+2] == 'id'):
            return True
    return False
```

```python
test_string_1 = "This is a test string with di and id."  
test_string_2 = "No in this string!"

# Перевірка для першої стрічки
result_1 = func_6_d(test_string_1)
print(f"Результат перевірки для стрічки 1: {result_1}")

# Перевірка для другої стрічки
result_2 = func_6_d(test_string_2)
print(f"Результат перевірки для стрічки 2: {result_2}")
```
